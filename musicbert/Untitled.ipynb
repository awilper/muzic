```python
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
layer = 2
weights = np.load("weights"+str(layer)+".npy")
old_weights = np.load("old_weights"+str(layer)+".npy")
```

```python
w_start = 0
w_size = 100
multi = 1
old_window = old_weights[multi,w_start:w_start+w_size,w_start:w_start+w_size]
window = weights[multi,w_start:w_start+w_size,w_start:w_start+w_size]
plt.matshow((window+0.0001)/(old_window+0.0001))
plt.matshow(np.log(window+0.01))
plt.matshow(np.log(old_window+0.01))
```
